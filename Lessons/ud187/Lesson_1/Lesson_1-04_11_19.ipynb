```
import numpy as np
```

```
an_arr = np.array(["Hello", "World"])
an_arr = np.append(an_arr, "!")
print(an_arr)

for element in an_arr:
  print(element)

for index, element in enumerate(an_arr):
  print(index, "--", element)

print(type(an_arr))
```

```
arr_1 = np.array([0, 1, 100, 5, -2, -30])
print(arr_1)

print("Max : {}".format(np.max(arr_1)))
print("Min : {}".format(np.min(arr_1)))
print("Max index : {}".format(np.argmax(arr_1)))
print("Min index : {}".format(np.argmin(arr_1)))
print("Average : {}".format(np.average(arr_1)))
```

```
print(type(an_arr), "  ", type(an_arr[0]))
print(type(arr_1), "  ", type(arr_1[0]))
```

```
rand_arr_1 = np.random.rand(3, 4)
print(rand_arr_1, "\n")

for row in rand_arr_1:
  print()
  print(row)
  for col in row:
    print(col)
  
print("\nShape of an_arr : {}".format(an_arr.shape))
print("Shape of arr_1 : {}".format(arr_1.shape))
print("Shape of rand_arr_1 : {}".format(rand_arr_1.shape))
```
