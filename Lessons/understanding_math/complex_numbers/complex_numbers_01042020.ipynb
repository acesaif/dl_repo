```
def random_complex_generator(nums=1, set_of=1, flatten=False):
  """This function generates random complex numbers.
  
  Parameters
  ----------
    `nums` : determines the number of complex numbers to be generated.
    `set_of` : specifies how many sets of those complex numbers are required.
    `flatten` : flattens the multi-dimensional array to single dimention.
    
  Returns
  -------
    `set_complexes` : a numpy array either flattened or multi-dimensional 
                      in nature filled with complex numbers"""
  
  set_complexes = [[complex(random.randint(a=-20, b=20), 
                            random.randint(a=-20, b=20)) for i in range(nums)] 
                   for s in range(set_of)]
  set_complexes = np.array(set_complexes)
  if flatten:
    set_complexes = set_complexes.flatten()
    return set_complexes
  else:
    return set_complexes
```

### Arithmetic Operations in Complex Numbers $(a + ib)$

Let $z_1 = (a + ib)$ and $z_2 = (c + id)$ be two complex numbers then, 

* Addition $: \ z_1 + z_2 = (a+c) + i(b+d)$

* Substraction $: \ z_1 - z_2 = (a-c) + i(b-d)$

* Multiplication $: \ z_1z_2 = (ac-bd) + i(ad+bc)$

* Division $: \ \frac{z_1}{z_2} = \frac{(a+ib)}{(c+id)} = \frac{(a+ib)(c-id)}{(c+id)(c-id)} = \frac{(ac+bd)}{(c^2+d^2)} +i\frac{(bc-ad)}{(c^2+d^2)}$

```
def simple_complex_guide(how_many=2, rand_c_gen=True, operation=None):
  """This function generates random complex number(s) and performs 
  arithmetic operations as per specifications and also visualizes the results.
  
  Parameter
  ---------
    `how_many` : default value is `2`, which means it produces 
                 `2` complex numbers. It can produce any number of
                 complex numbers as per parameterization.
    `rand_c_gen` : a switch to generate randomized numbers.
    `operation` : default value is `None`, user specifies the operation, 
                  strictly needs to be 'add', 'sub', 'mul', and 'div' for 
                  `+`, `-`, `*` and `/` respectively. It is case insensitive.
  
  Returns
  -------
  NA"""

  if how_many > 1:
    if rand_c_gen:
      c_num_list = [complex(random.randint(a=-10, b=10), 
                            random.randint(a=-10, b=10)) for c in range(how_many)]
    else:
      c_num_list = list()
      for c in range(how_many):
        print("Complex Number : {}".format(c+1))
        temp_real = int(input("\tPlease enter the real part : "))
        temp_imag = int(input("\tPlease enter the imag part : "))
        c_num_list.append(complex(temp_real, temp_imag))
    
    c_num_list = np.array(c_num_list)
    print(c_num_list)
  
  if operation == None:
    reals = np.array([c.real for c in c_num_list])
    imags = np.array([c.imag for c in c_num_list])

    plt.figure(figsize=(8, 6))
    plt.title("Argand Plane")

    plt.axhline(y=0, c='k', ls='--', lw=0.7)
    plt.axvline(x=0, c='k', ls='--', lw=0.7)

    plt.scatter(x=reals, y=imags, c='magenta', label='Given', marker='x')

    plt.xlim([-(max(np.abs(reals))*2), max(np.abs(reals))*2])
    plt.ylim([-(max(np.abs(imags))*2), max(np.abs(imags))*2])
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.legend(loc='center left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
  else:
    reals = np.array([c.real for c in c_num_list])
    imags = np.array([c.imag for c in c_num_list])

    if operation.lower() == 'add' or operation == '+':
      operated_complex = c_num_list[0]
      for c_num in range(1, len(c_num_list)):
        operated_complex += c_num_list[c_num]
      print("Operation (+) : {}".format(operated_complex))
    elif operation.lower() == 'sub' or operation == '-':
      operated_complex = c_num_list[0]
      for c_num in range(1, len(c_num_list)):
        operated_complex -= c_num_list[c_num]
      print("Operation (-) : {}".format(operated_complex))
    elif operation.lower() == 'mul' or operation == '*':
      operated_complex = c_num_list[0]
      for c_num in range(1, len(c_num_list)):
        operated_complex *= c_num_list[c_num]
      print("Operation (*) : {}".format(operated_complex))
    elif operation.lower() == 'div' or operation == '/':
      operated_complex = c_num_list[0]
      for c_num in range(1, len(c_num_list)):
        operated_complex /= c_num_list[c_num]
      op_real = round(operated_complex.real, 3)
      op_imag = round(operated_complex.imag, 3)
      operated_complex = complex(op_real, op_imag)
      print("Operation (/) : {}".format(operated_complex))
    
    mega_com_list = np.append(arr=c_num_list, values=operated_complex)

    mega_reals = np.array([c.real for c in mega_com_list])
    mega_imags = np.array([c.imag for c in mega_com_list])

    plt.figure(figsize=(8, 6))
    plt.title("Argand Plane")

    plt.axhline(y=0, c='k', ls='--', lw=0.7)
    plt.axvline(x=0, c='k', ls='--', lw=0.7)

    plt.scatter(x=reals, y=imags, c='magenta', label='Given', marker='x')
    plt.scatter(x=operated_complex.real, 
                y=operated_complex.imag, 
                c='green', marker='x', 
                label='Obtained after ' + operation.title())
    
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.xlim([-(max(np.abs(mega_reals))*2), max(np.abs(mega_reals))*2])
    plt.ylim([-(max(np.abs(mega_imags))*2), max(np.abs(mega_imags))*2])
    plt.legend(loc='center left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
```

### Complex Conjugate

If $z = (a + ib)$ is a complex number, then $\bar{z} = (a - ib)$ is its conjugate. Conjugate of complex of number helps in dividing a complex number.

```
def conjugate_complex(c_num, plot_it=False):
  """This function, given a complex number (a + ib), produces 
  its conjuate (a - ib) and visualizes the conjugate against the input number.
  
  Parameters
  ----------
    `c_num` : input complex number which is of the form (a + ib).
    `plot_it` : a switch to turn on plots or to turn off the same.
  
  Returns
  -------
    `conjugate` : a complex number (a - ib) is returned if, 
                  `plot_it` is turned off."""

  real_part = c_num.real
  imag_part = c_num.imag

  if imag_part > 0 or imag_part < 0:
    imag_part = -(imag_part)

  conjugate = complex(real_part, imag_part)

  if plot_it:
    print("Complex number z : {}".format(c_num))
    print("Conjugate of z   : {}".format(conjugate))

    plt.figure(figsize=(8, 6))
    plt.title("Argand Plane")

    plt.axhline(y=0, c='k', ls='--', lw=0.7)
    plt.axvline(x=0, c='k', ls='--', lw=0.7)
    
    plt.scatter(x=c_num.real, y=c_num.imag, 
                c='magenta', label="Given", marker='x')
    plt.scatter(x=conjugate.real, 
                y=conjugate.imag, 
                c='green', label='Conjugate', marker='x')
    
    plt.xlim([-(abs(c_num.real)+5), abs(c_num.real)+5])
    plt.ylim([-(abs(c_num.imag)+5), abs(c_num.imag)+5])
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.legend(loc='center left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
  else:
    return conjugate
```

### Modulus of Complex Number $(a + ib)$

* Modulus of a complex number $z=(a+ib)$ is the distance the number is away from the origin $O = (0 + i0)$.
* It is denoted as $r$.
* $r = |z| = \sqrt{a^2+b^2}$.
* Clearly, $r^2 = z\bar{z} = |z|^2 = (a+ib)(a-ib) = a^2+b^2$.

```
def modulus_complex(c_num, plot_it=False):
  """This function obtains modulus of the complex number and visualizes it.
  
  Parameters
  ----------
    `c_num` : input complex number which is of the form (a + ib).
    `plot_it` : a switch to turn on plots or to turn off the same.
  
  Returns
  -------
    `absolute` : the absolute or modulus of the complex number."""

  absolute = round(math.sqrt((c_num * conjugate_complex(c_num)).real), 3)

  if plot_it:
    print("Complex number : {}".format(c_num))
    print("Absolute (r) : {}".format(absolute))

    plt.figure(figsize=(8, 6))
    plt.title("Argand Plane")

    plt.axhline(y=0, c='k', ls='--', lw=0.7)
    plt.axvline(x=0, c='k', ls='--', lw=0.7)

    plt.plot([0, c_num.real], [0, c_num.imag], 
             ls='--', color='red', label='Modulus ($r$)', lw=0.7)
    
    plt.scatter(x=c_num.real, y=c_num.imag, 
                c='magenta', label="Complex number ($z_1$)", marker='x')
    plt.scatter(x=0, y=0, c='k', marker='o')

    plt.xlim([-(abs(c_num.real)+5), abs(c_num.real)+5])
    plt.ylim([-(abs(c_num.imag)+5), abs(c_num.imag)+5])
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.legend(loc='center left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
  else:
      return absolute
```

### Argument and Different forms of Complex Number $(a + ib)$

* Argument of the complex number is defined as the angle at which the complex number makes with positive Real axis.
* It is denoted as $\theta$.
* $\theta = \tan^{-1}(\frac{b}{a})$.
* It is also considered as the direction.

Given $\theta$ and $r$, we can obtain a complex number and can be plotted in Polar coordinate axis. Plots below are not drawn in polar axes.

* $z = r(\cos(\theta) + i\sin(\theta))$ this is Polar form of $C$.
* $z = re^{i\theta}$ is the Exponential form of $C$.
* $z = a + ib$ is the Rectangular form of $C$.

```
def argument_complex(c_num, plot_it=False, measure='rad'):
  """This functions obtains argument (angle or direction in polar form)
  and visualizes it.
  
  Parameters
  ----------
    `c_num` : input complex number which is of the form (a + ib).
    `plot_it` : a switch to turn on plots or to turn off the same.
    `measure` : default value is 'rad' stands for radians, but can be 
                toggled to 'deg' stands for degrees.
  
  Returns
  -------
    `argument` : the angle that a complex number makes with 
                 positive real axis, follows the 
                 range of [-3.142, 3.142] if radians 
                 or range of [-180, 180] if degrees."""

  try:
    if measure.lower() == 'rad':
      if c_num.real == 0 and c_num.imag == 0:
        argument = None
      elif c_num.real == 0 and c_num.imag > 0:
        argument = 1.570
      elif c_num.real == 0 and c_num.imag < 0:
        argument = -1.570
      else:
        try:
          argument = math.atan(c_num.imag/c_num.real)
          if c_num.real > 0 and c_num.imag >= 0:   # Q1
            argument += 0.0
          elif c_num.real < 0 and c_num.imag >= 0: # Q2
            argument += 3.142
          elif c_num.real < 0 and c_num.imag <= 0: # Q3
            argument -= 3.142
          elif c_num.real > 0 and c_num.imag <= 0: # Q4
            argument -= 0.0
        except:
          argument = argument
      argument = round(argument, 3)
    elif measure.lower() == 'deg':
      if c_num.real == 0 and c_num.imag == 0:
        argument = None
      elif c_num.real == 0 and c_num.imag > 0:
        argument = 90.0
      elif c_num.real == 0 and c_num.imag < 0:
        argument = -90.0
      else:
        try:
          argument = math.atan(c_num.imag/c_num.real) * 57.2958
          if c_num.real > 0 and c_num.imag >= 0:   # Q1
            argument += 0.0
          elif c_num.real < 0 and c_num.imag >= 0: # Q2
            argument += 180.0
          elif c_num.real < 0 and c_num.imag <= 0: # Q3
            argument -= 180.0
          elif c_num.real > 0 and c_num.imag <= 0: # Q4
            argument -= 0.0
        except:
          argument = argument
      argument = round(argument, 3)
  except:
    argument = 'NA'
  
  if plot_it:
    print("Complex number : {}".format(c_num))
    print('Argument (theta) : {}'.format(argument))
    
    plt.figure(figsize=(8, 6))
    plt.title("Argand Plane")

    plt.axhline(y=0, c='k', ls='--', lw=0.7)
    plt.axvline(x=0, c='k', ls='--', lw=0.7)

    plt.plot([0, c_num.real], [0, c_num.imag], 
             ls='--', color='red', label='Modulus ($r$)', lw=0.7)
    
    plt.scatter(x=c_num.real, y=c_num.imag, 
                c='magenta', label="Complex number ($z_1$)", marker='x')
    plt.scatter(x=0, y=0, c='k', marker='o')

    plt.xlim([-(abs(c_num.real)+5), abs(c_num.real)+5])
    plt.ylim([-(abs(c_num.imag)+5), abs(c_num.imag)+5])
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.legend(loc='center left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
  else:
      return argument
```

```
def rectangular_to_polar(c_num):
  """This function converts the complex number of (a + ib) 
  into polar form r(cos(0) + isin(0)).
  
  Parameter
  ---------
    `c_num` : input complex number which is of the form (a + ib).
  
  Returns
  -------
    `polar_complex` : a polar form of the inputted complex number."""

  modulus = modulus_complex(c_num=c_num)
  degress = argument_complex(c_num=c_num, measure='rad')
  a_complex = round((modulus * math.cos(degress)), 3)
  b_complex = round((modulus * math.sin(degress)), 3)

  polar_complex = complex(a_complex, b_complex)
  return polar_complex
```

```
def polar_to_rectangular(angle, modulus):
  """This function obtains a rectugular form of complex number 
  given the angle `0` and modulus `r`.
  
  Parameters
  ----------
    `angle` : the argument of the imminent resultant complex number.
    `modulus` : the distance from the origin of the 
                imminent resultant complex number.
  
  Returns
  -------
    `rectangular_complex` : an obtained rectangular complex number."""

  a_complex = round((modulus * math.cos(angle)), 3)
  b_complex = round((modulus * math.sin(angle)), 3)

  rectangular_complex = complex(a_complex, b_complex)
  return rectangular_complex
```
